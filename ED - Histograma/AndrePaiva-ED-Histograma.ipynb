## 1 - Importação de dependências e configurações

```python
import numpy as np
import matplotlib.pyplot as plt
import PIL
```

## 2 - Abrindo a imagem

A imagem é aberta utilizando a biblioteca `PIL` e convertida em `np.array`.
O histograma é calculado e salvo na variável `h`, o valor *argmax* (bin com maior valor de contagem) é salvo na variável `m` e então o valor de `m` é exibido como saída.

```python
with PIL.Image.open('castle.jpg') as im:
    picture = np.asarray(im)

h = np.histogram(picture, bins=256, range=(0, 255))[0]
m = h.argmax()
print(m)
```

## 3 - Plotando histograma

O histograma é plotado diretamente da matriz da imagem utilizando a função `histplot` da biblioteca `seaborn`.

```python
plt.hist(picture.flatten(), bins=256, range=(0, 255))
plt.show()
```

## 4 - Binarização da imagem

Baseado no *plot* exibido acima, foi escolhido um valor de *threshold* (salvo em `t`) que foi aplicado à imagem. A imagem binarizada é exibida como saída.

```python
t = 175
bin_pic = picture > t
plt.imshow(bin_pic, cmap = 'Greys_r')
plt.show()
```
