## 2 - Abrindo a imagem

Primeiramente, é importada a imagem original com representação numérica em `double`.

```python
# Leitura da imagem original
targets = Image.Read("yellowtargets.png", dtype='double', gamma='sRGB')
targets.disp()
```

## 3 - Plotando o plano `xy` de cromaticidade

```python
plot_chromaticity_diagram("xy")
```

## 4 - Aquisição da imagem no plano `xy`

```python
ab = targets.colorspace("L*a*b*").plane("a*:b*")
```

```python
ab.plane('a*:').disp()
```

```python
ab.plane('b*:').disp()
```

## 5 - Aplicação do algoritmo K-means

```python
labels, centroides, residual = ab.kmeans_color(k=2, seed=0)
labels = (labels == 0)
```
