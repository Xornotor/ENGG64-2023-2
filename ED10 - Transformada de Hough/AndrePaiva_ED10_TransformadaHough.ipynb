## 1 - Importação de dependências e configurações

Para esta atividade é utilizada a `machinevisiontoolbox` do Peter Corke e a biblioteca `matplotlib` para exibição de imagens.

```python
from machinevisiontoolbox.base import *
from machinevisiontoolbox import *
from spatialmath.base import *
from spatialmath import *
from matplotlib import pyplot as plt
```

## 2 - Abrindo a imagem

Primeiramente, é importada a imagem e é feita a binarização.

```python
# Leitura da imagem original
church = Image.Read("church.png", mono='True')
church.disp()
```

## 3 - Executando a Transformada de Hough

Foi efetuada a Transformada de Hough e selecionadas as linhas com no mínimo 60 votos.

```python
edges = church.canny()
h = edges.Hough()
lines = h.lines_p(60, minlinelength=200, maxlinegap=5, seed=0)
with plt.ioff():
    church.disp();
    h.plot_lines(lines, 'r--')
    plt.show()
```
