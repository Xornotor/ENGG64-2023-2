## 1 - Importação de dependências e configurações

Para esta atividade é utilizada a `machinevisiontoolbox` do Peter Corke e a biblioteca `matplotlib` para exibição de imagens.

```python
from machinevisiontoolbox import *
from matplotlib import pyplot as plt
```

## 2 - Abrindo a imagem (`uint8`)

Primeiramente, é importada a imagem original com representação numérica em `uint8`.

```python
# Leitura da imagem original
flowers = Image.Read("flowers9.png", dtype='uint8')
```

## 3 - Plotando Histogramas

```python
fig, ax = plt.subplots(3, figsize=(10, 8))
r_hist = ax[0].hist(flowers.image[:, :, 0].flatten(), color='red', range=(0,255), bins=255)
g_hist = ax[1].hist(flowers.image[:, :, 1].flatten(), color='green', range=(0,255), bins=255)
b_hist = ax[2].hist(flowers.image[:, :, 0].flatten(), color='blue', range=(0,255), bins=255)
plt.show()
```

## 4 - Plotando os planos de cor

```python
# Plot
fig, ax = plt.subplots(2, 2, figsize=(6, 8))
rgb = ax[0][0].imshow(flowers.image)
ax[0][0].set_title('RGB')
r = ax[0][1].imshow(flowers.image[:, :, 0], cmap='gray')
ax[0][1].set_title('R')
g = ax[1][0].imshow(flowers.image[:, :, 1], cmap='gray')
ax[1][0].set_title('G')
b = ax[1][1].imshow(flowers.image[:, :, 2], cmap='gray')
ax[1][1].set_title('B')
plt.show()
```
